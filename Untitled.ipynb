```python
df.columns=['State', 'Total Cases', 'Active', 'Discharged', 'Deaths',
       'Active Ratio', 'Discharge Ratio', 'Death Ratio']
```

```python
df
```

```python
df.groupby('State')['Active Ratio'].max().sort_values(ascending=False)
```

```python
plt.figure(figsize=(10,4))
plt.plot(df['State'],df['Active Ratio'],color='red',marker='o',markersize=10)
plt.xticks(rotation=90,fontsize=11)
plt.yticks(fontsize=13.5)
plt.xlabel('State',fontsize=15)
plt.ylabel('Active Ratio',fontsize=15)
plt.title('Line plot for Active Ratio',fontsize=15)
plt.show()
```

```python
#total cases in the state
total_cases_state=df.groupby('State')['Total Cases'].max().sort_values(ascending=False).reset_index()
total_cases_state
```

```python
# top 10 total cases in each state
plt.figure(figsize=(15,4))
plt.xticks(rotation=75)
sns.barplot(x='State',y='Total Cases',data=total_cases_state.head(10))
plt.xlabel('State',fontsize=15)
plt.ylabel('Total Cases',fontsize=15)
plt.title('Total Cases in each State',fontsize=20)
plt.show()
```

```python
# last 10 tatal cases in each state
plt.figure(figsize=(15,4))
plt.xticks(rotation=75)
sns.barplot(x='State',y='Total Cases',data=total_cases_state.tail(10))
plt.xlabel('State',fontsize=15)
plt.ylabel('Total Cases',fontsize=15)
plt.title('Total Cases in each State',fontsize=20)
plt.show()
```

```python
# pie chart for totsl cases in the states
# plt.figure(figsize=(8,8))
L=[0,0,0,0,0.5]
plt.pie(x='Total Cases',data=total_cases_state.head(),labels='State',radius=0.7,explode=L,shadow=True,startangle=80)
plt.title('Active cases in top 5 state in india')
# plt.legend(loc='center left')
```
